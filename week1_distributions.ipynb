To solve this issue, we use data types to determine how computers treat these objects, ensuring that we translates the data appropriately.

## Section 2: Statistics, distributions, and sensitivity analysis

This section contains some definitions and example calculations of statistical measures. We will use synthetic data relating to rolling a fair 6-sided die to illustrate each concept.

### Mean

#### Expected Value

**Definition**: The expected value of a random variable is a measure of the central tendency. It is calculated as the weighted average of all possible values. For discrete variables, it's the sum of the product of each value and its probability.

```python
# Example: Expected value of a fair die roll
values = np.array([1, 2, 3, 4, 5, 6])
probabilities = np.repeat(1/6, 6)  # Equal probability for a fair die = [1/6, ..., 1/6]
expected_value = np.sum(values * probabilities)
print(f"Expected Value: {expected_value}")
```

#### Sample Mean


**Definition**: The sample mean is the average value of a sample and is a practical estimate of the population mean or expected value.

```python
# Example: Sample mean from random die rolls
sample = np.random.choice(values, size=100, p=probabilities)
sample_mean = np.mean(sample)
print(f"Sample Mean: {sample_mean}")
```

#### Example: sample size and mean

Below is a plot of sample means with different sample sizes. The red horizontal line is the expected value.

```python
# Make sure widgets work in Colab
# !pip install ipywidgets
# from google.colab import output
# output.enable_custom_widget_manager()

# Define the base function for means visualization
def get_sample_means(sample_size=10, num_samples=1000):
    sample_means = np.zeros(num_samples)
    for i in range(num_samples):
        sample = np.random.choice(values, size=sample_size, p=probabilities)
        sample_means[i] = np.mean(sample)
    return sample_means

def dice_mean_visualization(dice_per_sample=10, num_samples=1000, num_bins=30):
    """
    Base function to generate and visualize the sampling distribution of dice roll means
    """
    np.random.seed(42)
    sample_means = get_sample_means(dice_per_sample, num_samples)
    population_mean = np.mean(sample_means)
    expected_mean = 3.5  # (1+2+3+4+5+6)/6

    fig, ax = plt.subplots(figsize=(8, 6))
    ax.hist(sample_means, bins=num_bins,
            alpha=0.75,
            color='skyblue',
            edgecolor='black')

    ax.axvline(x=expected_mean, color='red', linestyle='--', linewidth=2,
              label=f'Expected Mean: {expected_mean}')
    ax.axvline(x=population_mean, color=(0, 1, 0, 0.5), linestyle='-', linewidth=2,
              label=f'Observed Mean: {population_mean:.4f}')
    ax.set_title(f'Distribution of sample means\n({dice_per_sample} dice per sample, {num_samples} samples)',
                fontsize=14)
    ax.set_xlabel('Sample Mean Value', fontsize=12)
    ax.set_ylabel('Frequency', fontsize=12)
    ax.grid(True, alpha=0.3)
    ax.legend(fontsize=10)

    textstr = '\n'.join((
        f'Rolls per sample: {dice_per_sample}',
        f'Number of samples: {num_samples}',
        f'Expected Mean: {expected_mean}',
        f'Observed Mean: {population_mean:.4f}',
        f'Difference: {population_mean - expected_mean:.4f}'
    ))

    props = dict(boxstyle='round', facecolor='wheat', alpha=0.5)
    ax.text(0.05, 0.95, textstr, transform=ax.transAxes, fontsize=10,
        verticalalignment='top', bbox=props)

    plt.tight_layout()
    plt.show()
    print("=== MEAN STATISTICS ===")
    print(f"Rolls per sample: {dice_per_sample}")
    print(f"Number of samples: {num_samples}")
    print(f"Expected Mean: {expected_mean}")
    print(f"Observed Mean: {population_mean:.4f}")
    print(f"Difference: {population_mean - expected_mean:.4f}")

# Create the interactive widget
@widgets.interact_manual(
    dice_per_sample=widgets.IntSlider(
        min=1,
        max=100,
        step=1,
        value=10,
        description='Dice per sample:',
        style={'description_width': 'initial'}
    ),
    num_samples=widgets.IntSlider(
        min=10,
        max=10000,
        step=10,
        value=1000,
        description='Number of samples:',
        style={'description_width': 'initial'}
    ),
    num_bins=widgets.IntSlider(
        min=10,
        max=50,
        step=5,
        value=30,
        description='Number of bins:',
        style={'description_width': 'initial'}
    )
)
def interactive_dice_mean(dice_per_sample=10, num_samples=1000, num_bins=30):
    """Interactive wrapper for the base function"""
    dice_mean_visualization(dice_per_sample, num_samples, num_bins)
```

### Variance

**Definition**: Variance measures the dispersion of a set of data points around their mean value. It's defined as the average of the squared differences from the Mean.

#### Dice variance

```python
# Example: Variance of a fair die roll
# Var[x] = E[(X-E(X))^2]
values = np.array([1, 2, 3, 4, 5, 6])
variance = np.sum((values - expected_value)**2 * probabilities)
print(f"Variance: {variance}")
```

```python
# Var[X] = E[X^2] - (E[X])^2
e_x = 3.5
values_square = [i ** 2 for i in range(1, 7)]
e_x2 = np.sum(values_square * probabilities)
var = e_x2 - e_x ** 2
var
```

#### Standard error of the sample mean

Definition: The standard error of the mean (SEM) measures how far the sample mean of the data is likely to be from the true population mean. It's calculated as the standard deviation of the sample divided by the square root of the sample size.

```python
# Example: Standard error of the sample mean from random die rolls (Var[X_bar])
# SEM = dice std/sqrt(number of samples)
standard_error = np.std(sample, ddof=1) / np.sqrt(len(sample))
print(f"Standard Error: {standard_error}")
```

#### Example: sample size and standard deviation

```python
# Make sure widgets work in Colab
# !pip install ipywidgets
# from google.colab import output
# output.enable_custom_widget_manager()

probabilities = np.repeat(1/6, 6)
values = np.array([1, 2, 3, 4, 5, 6])
def get_sample_means_std(sample_size=10, num_samples=1000):
    sample_statistics = {
        "mean": [],
        "variance": [],
    }
    for i in range(num_samples):
        sample = np.random.choice(values, size=sample_size, p=probabilities)
        sample_statistics["mean"].append(np.mean(sample))
        sample_statistics["variance"].append(np.var(sample))
    return sample_statistics

def dice_mean_std_visualization(dice_per_sample=10, num_samples=1000, num_bins=30):
    """
    Base function to generate and visualize the sampling distribution of dice roll means
    """
    np.random.seed(42)
    sample_statistics = get_sample_means_std(dice_per_sample, num_samples)
    population_mean = np.mean(sample_statistics["mean"])
    population_std = np.std(sample_statistics["mean"])
    expected_sample_mean = 3.5
    expected_sample_std = np.sqrt(np.sum((np.array([1,2,3,4,5,6]) - expected_sample_mean)**2 * probabilities)) # std for sample
    sample_std = np.mean(list(map(np.sqrt, sample_statistics["variance"])))
    expected_population_mean = expected_sample_mean
    expected_population_std = expected_sample_std / np.sqrt(dice_per_sample) # std of population mean (SEM) = sample std/sqrt(dice per sample)

    fig, ax = plt.subplots(figsize=(8, 6))
    ax.hist(sample_statistics["mean"], bins=num_bins,
            alpha=0.75,
            color='skyblue',
            edgecolor='black')

    ax.axvline(x=expected_population_mean, color='red', linestyle='--', linewidth=2,
              label=f'Expected Mean: {expected_population_mean:.4f}')
    ax.axvline(x=population_mean, color=(0, 1, 0, 0.5), linestyle='-', linewidth=2,
              label=f'Observed Mean: {population_mean:.4f}')
    ax.axvline(x=population_mean + 2*population_std, color='purple', linestyle=':', linewidth=1,
              label=f'+2 SD')
    ax.axvline(x=population_mean - 2*population_std, color='purple', linestyle=':', linewidth=1,
              label=f'-2 SD')

    ax.set_title(f'Distribution of sample means and standard deviation\n({dice_per_sample} dice per sample, {num_samples} samples)',
                fontsize=14)
    ax.set_xlabel('Sample Mean Value', fontsize=12)
    ax.set_ylabel('Frequency', fontsize=12)
    ax.set_xlim(0.9, 6.1)
    ax.grid(True, alpha=0.3)
    ax.legend(fontsize=10)

    textstr = '\n'.join((
        f'Expected Mean: {expected_population_mean}',
        f'Observed Mean: {population_mean:.4f}',
        f'Expected Population Std: {expected_population_std:.4f}',
        f'Observed Population Std: {population_std:.4f}',
        f"Expected Sample Std: {expected_sample_std:.4f}",
        f'Observed Sample Std: {sample_std:.4f}',
    ))

    props = dict(boxstyle='round', facecolor='wheat', alpha=0.5)
    ax.text(0.05, 0.95, textstr, transform=ax.transAxes, fontsize=10,
        verticalalignment='top', bbox=props)

    plt.tight_layout()
    plt.show()
    print("=== MEAN STATISTICS ===")
    print(f"Expected Mean: {expected_population_mean}")
    print(f"Observed Mean: {population_mean:.4f}")
    print(f"Expected Std: {expected_population_std:.4f}")
    print(f"Observed Sample Std: {population_std:.4f}")

# Create the interactive widget
@widgets.interact_manual(
    dice_per_sample=widgets.IntSlider(
        min=1,
        max=100,
        step=1,
        value=10,
        description='Dice per sample:',
        style={'description_width': 'initial'}
    ),
    num_samples=widgets.IntSlider(
        min=10,
        max=10000,
        step=10,
        value=1000,
        description='Number of samples:',
        style={'description_width': 'initial'}
    ),
    num_bins=widgets.IntSlider(
        min=10,
        max=50,
        step=5,
        value=30,
        description='Number of bins:',
        style={'description_width': 'initial'}
    )
)
def interactive_dice_mean(dice_per_sample=10, num_samples=1000, num_bins=30):
    """Interactive wrapper for the base function"""
    dice_mean_std_visualization(dice_per_sample, num_samples, num_bins)
```

## Section 3: Data exploration: Understand a dataset
---


In our recent class, we covered the topics of expected values and variances as ways to describe data. To get a better feel for what these terms mean, we're going to create some example data. This practical exercise will help us understand the implications of these measures.

#### Distribution plotter

```python
# Function to generate data and plot different distributions
def plot_distribution_data(distribution_type='normal', param1=5, param2=1, num_samples=100):
    """
    Plot synthetic data and theoretical probability density function for different distributions

    Parameters:
    - distribution_type: Type of distribution to plot
    - param1: First parameter (meaning depends on distribution)
    - param2: Second parameter (meaning depends on distribution)
    - num_samples: Number of samples to generate
    """
    # Setup figure and axes
    fig, axs = plt.subplots(1, 2, figsize=(15, 6))

    # Generate data based on selected distribution
    if distribution_type == 'normal':
        mean, std = param1, np.sqrt(param2)
        data = np.random.normal(mean, std, num_samples)
        x = np.linspace(mean - 4*std, mean + 4*std, 1000)
        y = stats.norm.pdf(x, mean, std)
        param1_name = "Mean"
        param2_name = "Variance"
        dist_title = f"Normal Distribution (μ={mean}, σ²={param2})"

    elif distribution_type == 'uniform':
        low, high = param1, param1 + param2  # param1=min, param2=range
        data = np.random.uniform(low, high, num_samples)
        x = np.linspace(low - 0.5, high + 0.5, 1000)
        y = stats.uniform.pdf(x, low, param2)
        param1_name = "Lower Bound"
        param2_name = "Range"
        dist_title = f"Uniform Distribution (a={low}, b={high})"

    elif distribution_type == 'bernoulli':
        p = param1 / 10  # Scale to 0-1
        data = np.random.binomial(1, p, num_samples)  # Bernoulli is binomial with n=1
        x = np.array([0, 1])
        y = [1-p, p]
        param1_name = "Success Prob p (÷10)"
        param2_name = "Unused"
        dist_title = f"Bernoulli Distribution (p={p})"

    elif distribution_type == 'binomial':
        n, p = int(param1), param2 / 10  # Scale to 0-1
        data = np.random.binomial(n, p, num_samples)
        x = np.arange(0, n+1)
        y = stats.binom.pmf(x, n, p)
        param1_name = "Trials n"
        param2_name = "Success Prob p (÷10)"
        dist_title = f"Binomial Distribution (n={n}, p={p})"

    elif distribution_type == 'poisson':
        lam = param1
        data = np.random.poisson(lam, num_samples)
        x = np.arange(0, max(20, int(lam*3)))
        y = stats.poisson.pmf(x, lam)
        param1_name = "Lambda λ"
        param2_name = "Unused"
        dist_title = f"Poisson Distribution (λ={lam})"

    # Plot the synthetic data
    axs[0].scatter(range(len(data)), data, alpha=0.7)
    axs[0].set_title(f'Synthetic Data Samples ({distribution_type.capitalize()})', fontsize=16)
    axs[0].set_xlabel('Sample Index', fontsize=14)
    axs[0].set_ylabel('Value', fontsize=14)
    axs[0].grid(True, alpha=0.3)

    # Plot the probability density function
    if distribution_type in ['bernoulli', 'binomial', 'poisson']:
        # For discrete distributions, use bar chart
        axs[1].bar(x, y, alpha=0.7, width=0.8)
        x_label = "Outcome"
        y_label = "Probability Mass"
    else:
        # For continuous distributions, use line plot
        axs[1].plot(x, y, linewidth=2)
        x_label = "Value"
        y_label = "Probability Density"

    axs[1].set_title(dist_title, fontsize=16)
    axs[1].set_xlabel(x_label, fontsize=14)
    axs[1].set_ylabel(y_label, fontsize=14)
    axs[1].grid(True, alpha=0.3)

    # Display statistics
    stats_dict = {}
    if distribution_type == 'normal':
        stats_dict = {
            'Mean (Expected)': mean,
            'Mean (Sample)': np.mean(data),
            'Variance (Expected)': param2,
            'Variance (Sample)': np.var(data),
            'Std Dev (Expected)': std,
            'Std Dev (Sample)': np.std(data)
        }
    elif distribution_type == 'uniform':
        stats_dict = {
            'Mean (Expected)': (low + high) / 2,
            'Mean (Sample)': np.mean(data),
            'Variance (Expected)': param2**2 / 12,
            'Variance (Sample)': np.var(data),
            'Min (Expected)': low,
            'Min (Sample)': np.min(data),
            'Max (Expected)': high,
            'Max (Sample)': np.max(data)
        }
    elif distribution_type == 'bernoulli':
        stats_dict = {
            'Mean (Expected)': p,
            'Mean (Sample)': np.mean(data),
            'Variance (Expected)': p * (1-p),
            'Variance (Sample)': np.var(data)
        }
    elif distribution_type == 'binomial':
        stats_dict = {
            'Mean (Expected)': n * p,
            'Mean (Sample)': np.mean(data),
            'Variance (Expected)': n * p * (1-p),
            'Variance (Sample)': np.var(data)
        }
    elif distribution_type == 'poisson':
        stats_dict = {
            'Mean (Expected)': lam,
            'Mean (Sample)': np.mean(data),
            'Variance (Expected)': lam,
            'Variance (Sample)': np.var(data)
        }

    # Add statistics box to the plot
    stat_text = '\n'.join([f'{k}: {v:.4f}' for k, v in stats_dict.items()])
    props = dict(boxstyle='round', facecolor='wheat', alpha=0.5)
    axs[0].text(0.05, 0.95, stat_text, transform=axs[0].transAxes, fontsize=10,
        verticalalignment='top', bbox=props)

    # Print statistics
    print(f"=== {distribution_type.upper()} DISTRIBUTION STATISTICS ===")
    for k, v in stats_dict.items():
        print(f"{k}: {v:.4f}")

    plt.tight_layout()
    plt.show()

    # Return parameter names for widget labels
    return param1_name, param2_name
```

 #### Update parameter range helper function

```python
# Function to update parameter ranges and labels based on selected distribution
def update_parameter_ranges(change):
    distribution = change.new

    if distribution == 'normal':
        param1_slider.min = 0
        param1_slider.max = 20
        param1_slider.step = 0.5
        param1_slider.value = 5

        param2_slider.min = 0.1
        param2_slider.max = 10
        param2_slider.step = 0.1
        param2_slider.value = 1

    elif distribution == 'uniform':
        param1_slider.min = 0
        param1_slider.max = 50
        param1_slider.step = 1
        param1_slider.value = 0

        param2_slider.min = 1
        param2_slider.max = 50
        param2_slider.step = 1
        param2_slider.value = 10

    elif distribution == 'bernoulli':
        param1_slider.min = 0
        param1_slider.max = 10
        param1_slider.step = 0.1
        param1_slider.value = 5

        param2_slider.min = 0
        param2_slider.max = 1
        param2_slider.step = 0.1
        param2_slider.value = 0

    elif distribution == 'binomial':
        param1_slider.min = 1
        param1_slider.max = 30
        param1_slider.step = 1
        param1_slider.value = 10

        param2_slider.min = 0
        param2_slider.max = 10
        param2_slider.step = 0.1
        param2_slider.value = 5

    elif distribution == 'poisson':
        param1_slider.min = 0.1
        param1_slider.max = 20
        param1_slider.step = 0.1
        param1_slider.value = 5

        param2_slider.min = 0
        param2_slider.max = 1
        param2_slider.step = 0.1
        param2_slider.value = 0

    # Run the function once to update parameter labels
    with output:
        output.clear_output(wait=True)
        param1_name, param2_name = plot_distribution_data(
            distribution_type=distribution,
            param1=param1_slider.value,
            param2=param2_slider.value,
            num_samples=samples_slider.value
        )
        param1_label.value = f'<div style="width:130px;">{param1_name}:</div>'
        param2_label.value = f'<div style="width:130px;">{param2_name}:</div>'
```

#### Interactive setup

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
from ipywidgets import widgets, Layout, HBox, VBox

# Create the interactive widgets
distribution_dropdown = widgets.Dropdown(
    options=[
        ('Normal', 'normal'),
        ('Uniform', 'uniform'),
        ('Bernoulli', 'bernoulli'),
        ('Binomial', 'binomial'),
        ('Poisson', 'poisson')
    ],
    value='normal',
    description='Distribution:',
    style={'description_width': 'initial'}
)

# Create parameter sliders with placeholder labels that will be updated
param1_label = widgets.HTML('<div style="width:130px;">Parameter 1:</div>')
param1_slider = widgets.FloatSlider(
    min=0, max=20, step=0.5, value=5,
    readout_format='.1f',
    layout=Layout(width='250px')
)

param2_label = widgets.HTML('<div style="width:130px;">Parameter 2:</div>')
param2_slider = widgets.FloatSlider(
    min=0, max=10, step=0.1, value=1,
    readout_format='.1f',
    layout=Layout(width='250px')
)

samples_slider = widgets.IntSlider(
    min=10, max=1000, step=10, value=100,
    description='Sample Size:',
    style={'description_width': 'initial'},
    layout=Layout(width='320px')
)

# Group the parameter sliders with their labels
param1_control = HBox([param1_label, param1_slider])
param2_control = HBox([param2_label, param2_slider])

# Create a manual button for running the simulation
run_button = widgets.Button(
    description='Run Simulation',
    button_style='primary',
    tooltip='Click to run the simulation with current parameters',
    layout=Layout(width='150px')
)

# Organize all controls
top_controls = HBox([distribution_dropdown, samples_slider])
bottom_controls = HBox([param1_control, param2_control, run_button])
all_controls = VBox([top_controls, bottom_controls])
```

#### Interactive plot

```python
# Create an output widget to display the plot
output = widgets.Output()

# Function to handle the run button click
def on_run_button_clicked(b):
    with output:
        output.clear_output(wait=True)
        param1_name, param2_name = plot_distribution_data(
            distribution_type=distribution_dropdown.value,
            param1=param1_slider.value,
            param2=param2_slider.value,
            num_samples=samples_slider.value
        )
        param1_label.value = f'<div style="width:130px;">{param1_name}:</div>'
        param2_label.value = f'<div style="width:130px;">{param2_name}:</div>'

# Connect the update function to the dropdown's value change
distribution_dropdown.observe(update_parameter_ranges, names='value')

# Connect the run button to its handler
run_button.on_click(on_run_button_clicked)

# Display the controls and output
display(all_controls)
display(output)

# Initialize with the default distribution
update_parameter_ranges(type('obj', (object,), {'new': distribution_dropdown.value}))
```

The code below calculates the pearson and spearman correlations.

```python
# Calculate Pearson's correlation
pearson_corr, _ = stats.pearsonr(body_mass, flipper_length)
print(f"Pearson's correlation: {pearson_corr:.3f}")

# Calculate Spearman's correlation
spearman_corr, _ = stats.spearmanr(body_mass, flipper_length)
print(f"Spearman's correlation: {spearman_corr:.3f}")

# Calculate covariance
cov = np.cov(body_mass, flipper_length)
print(f"covariance: {cov[0][1]:.3f}")
body_mean = np.mean(body_mass)
flipper_mean = np.mean(flipper_length)

covariance_cal = sum((x - body_mean) * (y - flipper_mean) for x, y in zip(body_mass, flipper_length)) / (len(body_mass) - 1)
print(f"cov_cal: {covariance_cal}")
```
